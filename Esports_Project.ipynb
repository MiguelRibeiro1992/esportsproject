### What is League of Legends?

*League of Legends* is a multiplayer online battle arena (MOBA) game in which the player controls a character ("champion") with a set of unique abilities.  
The game is played in a symmetrical map, divided by 3 lanes, where each half of the map belongs initially to each of the 2 teams competing.  
  

The map is composed by each team's base (which has a big crystal called "Nexus" and 3 small ones called "inhibitors"), multiple towers that prevent enemies to approach the bases by dealing damage and troops called "minions" that continuously spawn and push opposing troops to their base in order to advance and conquer the map.  
The final objective requires players to join forces with team mates and minions to outperform their lane opponents getting more resources like gold, for example (that allow the player to buy stronger items to equip), to destroy enemy towers and keep advancing until they reach enemy nexus and destroy it, achieving victory.

Released in 2009, this real-time moving chess-like game requires not only great strategic thinking but high mechanical skills and prior knowledge of the game's vast fundamentals (such as champions' abilities and movesets, item actives and passive effects, synergies and many others) to achieve victory.  
Because it is a team-based game, individuals play with four other players on their team (total of 5 players per team).  
Due to its highly team-dependent nature, it requires players to have synergy with their team mates and decide every play based on their team's current in-game state.

The ingredients were all present to conceive a League of Legends Esports scene, which has happened with astonishing international success.  
League of Legends rapidly became one of the most watched esports in the world, bringing competition to the international stage with world events, thousands of people to arenas/stadiums and millions to live-streaming platforms to watch the games.

### Variables explanation guide  
**Player**: In-game name which is unique to a player;  
**Opponent**: Unique asian team that G2 faced;  
**Role**: Position the player takes on the map;  
**Side**: In which side of the map G2 played (Blue - Left; Red - Right);  
**Champion**: Character the player chose to play in that game;  
**Kills, Deaths, Assists**: Part of player scoreboard;  
**KDA**: (Sum of kills)+(Sum of assists)/Total number of Deaths;  
**CS**: Creep Score, number of minions slain;  
**CSM**: Number of minions slains per minute;  
**Total_Gold**: Total gold generated by the player by the end of the match;  
**Vision_Score**: Total score generated by vision sentinels placed on the map by the player;  
**Total_Damage**: Total damage to enemy champions;  
**KP**: Kill participation (participation percentage in team's takedowns);  
**GD@15**: Gold difference compared to the opponent at 15 minute mark;  
**CSD@D**: Creep score difference compared to the opponent at 15 minute mark;  
**Towers**: Number of towers destroyed as a team in the game;  
**Dragons**: Number o dragons slain as a team in the game;  
**Barons**: Number of Barons slain as a team in the game;  
**Total_Gold_Difference**: Difference in total gold between teams by the end of the game.

```python
#Checking G2's wins and losses

total_wins = data['Result'].value_counts()['Win']
print(f"The number of wins is: {total_wins}")

total_losses = data['Result'].value_counts()['Loss']
print(f"The number of losses is: {total_losses}")

total_games = data['Result'].value_counts().sum()
print(f"The number of total games played: {total_games}")

#Checking the total winrate, rounding to 2 decimal places and converting into percentage
total_winrate = total_wins/total_games
total_winrate = total_winrate.round(decimals=2)
total_winrate = '{:.1%}'.format(total_winrate)

print(f"The winrate: {total_winrate}")
```

```python
#Removing space from 'DWG' element (method 1)
data['Opponent'] = data['Opponent'].apply(lambda x: x.strip())
data['Opponent'].unique()
```

```python
#Removing space from 'Loss' element (method 2)
data['Result'] = data['Result'].str.strip()
data['Result'].unique()
```
